# Creeer dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
avond_kcal= [737.5,500,537.5, 600]
lunch_kcal= [500,450,600,450,350.5]
```

# Maak boxplot

```python
data =(avond_kcal, lunch_kcal)
 
#fig = plt.figure(figsize =(10, 7))
 
# Creating plot
_ = plt.boxplot(data, labels=['diner', 'lunch'])
_ = plt.xlabel('Maaltijdsoort')
_ = plt.ylabel('Aanbevolen aantal Kcal')
_ = plt.title('Aanbevolen aantal Kcal per maaltijdsoort')
 
# show plot
plt.show()
```

# Krijg beeld van data

```python
print(statistics.median(avond_kcal))
print(statistics.median(lunch_kcal))
```

```python
print(np.mean(avond_kcal))
print(np.mean(lunch_kcal))
```

# Maak plot voor diner Kcal verdeling

```python
_ = plt.hist(avond_kcal, color = 'purple', bins = 3)
_ = plt.xlabel('Aanbevolen aantal Kcal')
_ = plt.ylabel('Aantal aanbevelingen')
_ = plt.title('Aanbevolen aantal Kcal voor het diner')
plt.show()
```
